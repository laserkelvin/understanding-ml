# Gaining an Intuition for Regularization

In machine learning, regularization is a approach used to combat high variance — in other words, the issue of your model learning to reproduce the data, rather than the underlying semantics about your problem. In an analogous way to humans learning, the idea is to construct your homework problems to test and build for knowledge, rather than simply route learning: for example, learning multiplication tables as opposed to learning _how_ to multiply.

This kind of phenomenon is especially prevalent in learning by neural networks — with great learning capacity comes a large likelihood for memorization, and it is up to us pracitioners to guide deep learning models into soaking up our problem, not our data. Many of you will have come across these methods in the past, and may have developed your own intuition for how different regularization methods affect the outcome. For those of you who don't (as I didn't before I wrote this!), this notebook provides a visual guide for neural networks are shaped by regularization.

## Create our "complex" data

To create our dataset for learning, I'm going to create some arbitrarily complex function that mimics periodic signals, but with some funkiness added to it. The function below implements the following equation:

$$ f(x) = \frac{\sin(A X)(1 - \cos(B X)}{\exp(C X)} $$

where $A, B, C$ are random numbers sampled from different Gaussian distributions. The effect of each of these values is to add lag between very similar functions, such that they add together randomly to generate _very_ different values of $f(x)$.

```julia
function create_data(data_length, normalize=true)
    X = collect(range(0., stop=4 * π, length=data_length));
    Y = (sin.(X .* rand(Normal(1., 0.5))))# .* (1. .- cos.(X .* rand(Normal(1., 0.5)))))# ./ exp.(X .* rand(Normal(0., 0.1)))
    # add white noise to the result
    Y .+= rand(Normal(0., 0.1), data_length)
    # flag to normalize the data into the range [0, 1]
    if normalize == true
        Y .-= minimum(Y)
        Y ./= (maximum(Y) - minimum(Y))
    end
    return X, Y
end
```

```julia
X, Y = create_data(1000, true);
```

```julia
plot(X, Y, label=false)
```

## Shuffling and splitting up into training/test sets

```julia
indices = collect(1:length(X));
indices = shuffle(indices);
```

```julia
shuffled_X = X[indices];
shuffled_Y = Y[indices];
```

```julia
train_X, test_X = splitobs(shuffled_X, at=0.7);
train_Y, test_Y = splitobs(shuffled_Y, at=0.7);
```

```julia
scatter(train_X, train_Y, label="Training set", alpha=0.7)
scatter!(test_X, test_Y, label="Test set", alpha=0.7)
```

## Defining our neural network model

We will be using a very simply neural network model to do its most core functionality: take an input value, apply a series of linear transformations, and generate an output. The parameters of our model, $\theta$, are going to hopefully learn to approximate our function $f(x)$, such that $f_\theta^*(x) \approx f(x)$.

```julia
function create_linear_model(n_hidden)
    model = Chain(
        Dense(1, n_hidden, initW=Flux.glorot_normal),
        Dense(n_hidden, n_hidden, initW=Flux.glorot_normal),
        Dense(n_hidden, n_hidden, initW=Flux.glorot_normal),
        Dense(n_hidden, n_hidden, initW=Flux.glorot_normal),
        Dense(n_hidden, 1)
    )
    return model
end
```

# Set up our training loop

```julia
function penalty(p)
    if p <= 0.
        lp_penalty = 0.
    else
        lp_penalty = norm(θ, p)
    end
    return lp_penalty
end
        

function loss(x, y, p, λ)
    J = Flux.mse(model(x), y) + (penalty(p) * λ)
    return J
end
```

# Train the model

```julia
function flattenall(a::AbstractArray)
    # Julia snippet that will flatten a nested array
    # https://gist.github.com/ivirshup/e9148f01663278ca4972d8a2d9715f72
    while any(x->typeof(x)<:AbstractArray, a)
        a = collect(flatten(a))
    end
    return a
end

function plot_parameter_pdf(θ)
    weights = flattenall([param for param in θ])
    weight_x = range(minimum(weights), stop=maximum(weights), length=1000);
    density = kde(weights, bandwidth=0.02);
    p_θ = pdf(density, weight_x);
    # Normalize probability
    p_θ ./= sum(p_θ)
    return weight_x, p_θ
end
```

```julia
function train_model(n_hidden, p, λ, num_epochs, η, train_data, test_data)
    # Create our fully connected model
    model = create_linear_model(n_hidden)
    # Define model parameters to learn
    θ = Flux.params(model)
    # Use gradient descent to train our model
    optimizer = ADAM(η)
    # arrays for tracking losses
    train_loss = zeros(num_epochs)
    test_loss = zeros(num_epochs)
    # Define our loss model; mean-squared-loss + regularization
    loss(x, y) = Flux.mse(model(x), y) + penalty(p) * λ
    # get data into correct orientation
    train_x, train_y = transpose(train_data[1]), transpose(train_data[2])
    test_x, test_y = transpose(test_data[1]), transpose(test_data[2])
    # training loop
    for epoch in 1:num_epochs
        # run through training set; vectorized
        J = loss(train_x, train_y)
        if epoch % 5 == 0
            @show epoch
            @show J
        end
        train_loss[epoch] = J
        # evaluate gradient of loss with respect to parameters θ
        grads = gradient(() -> loss(train_x, train_y), θ)
        # update θ according to ADAM optimizer with gradients
        Flux.Optimise.update!(optimizer, θ, grads)
        # perform validation
        test_loss[epoch] = loss(test_x, test_y)
    end
    return train_loss, test_loss, model, θ
end
```

```julia
train_loss, test_loss, model, θ = train_model(30, 0., 0., 50, 1e-3, (train_X, train_Y), (test_X, test_Y));
```

```julia
weight_x, p_θ = plot_parameter_pdf(θ)
plot(weight_x, p_θ)
```

```julia
plot(1:50, log.(train_loss), label="Training loss", lw=1.5)
plot!(1:50, log.(test_loss), label="Test loss", lw=1.5)
xlabel!("Epoch")
ylabel!("log MSE loss")
```
