# Gaining an Intuition for Regularization

In machine learning, regularization is a approach used to combat high variance — in other words, the issue of your model learning to reproduce the data, rather than the underlying semantics about your problem. In an analogous way to humans learning, the idea is to construct your homework problems to test and build for knowledge, rather than simply route learning: for example, learning multiplication tables as opposed to learning _how_ to multiply.

This kind of phenomenon is especially prevalent in learning by neural networks — with great learning capacity comes a large likelihood for memorization, and it is up to us pracitioners to guide deep learning models into soaking up our problem, not our data. Many of you will have come across these methods in the past, and may have developed your own intuition for how different regularization methods affect the outcome. For those of you who don't (as I didn't before I wrote this!), this notebook provides a visual guide for neural networks are shaped by regularization.

## Create our "complex" data

To create our dataset for learning, I'm going to create some arbitrarily complex function that mimics periodic signals, but with some funkiness added to it. The function below implements the following equation:

$$ f(x) = \frac{\sin(A X)(1 - \cos(B X)}{\exp(C X)} $$

where $A, B, C$ are random numbers sampled from different Gaussian distributions. The effect of each of these values is to add lag between very similar functions, such that they add together randomly to generate _very_ different values of $f(x)$.

```python
def create_dataset(length=1000, normalize=True):
    X = np.linspace(0., 4 * np.pi, length)
    A, B = np.random.normal(1., 0.5, 2)
    C = np.random.normal(0., 0.1)
    Y = (np.sin(X * A) * (1. - np.cos(X * B))) / np.exp(X * C)
#     Y = np.sin(X * A)
    # Add gaussian noise
    Y += np.random.normal(0., 0.1, Y.size)
    # Scale to range of [0, 1]
    if normalize:
        Y = (Y - Y.min()) / (Y.max() - Y.min())
    return X, Y
```

## Define our models

```python
class StandardUnit(nn.Module):
    """
    Standard stack of Linear + Activation + Dropout.
    This is just a convenience building block.
    """
    def __init__(self, n_in: int, n_out: int, dropout_p: int, activation=None):
        super().__init__()
        self.linear = nn.Linear(n_in, n_out)
        self.activation = activation
        self.dropout_p = dropout_p
    
    def forward(self, x: torch.Tensor):
        output = self.linear(x)
        output = F.dropout(output, self.dropout_p)
        if self.activation:
            output = self.activation(output)
        return output


class SimpleModel(nn.Module):
    """
    The full stack of our model. A single input/output value, and a
    user specified number of layers that are n_hidden wide.
    
    LeakyReLU activation is used for these to combat vanishing
    gradients. I had some issues previously where some functions
    were complex enough that anything else would simply refuse
    to fit.
    """
    def __init__(self, n_hidden, n_layers, dropout_p=0.):
        super().__init__()
        self.input_layer = nn.Linear(1, n_hidden)
        # This little portion generates a stack of fully connected layers
        # with LeakyReLU activation and dropouts between each layer.
        self.hidden_layers = nn.Sequential(
            *[StandardUnit(n_hidden, n_hidden, dropout_p, nn.LeakyReLU(0.03, inplace=True)) for _ in range(n_layers)]
        )
        # Values of our data are positive, and between 0 and 1. ReLU activation
        # on the output makes sure we don't have negative values.
        self.output_layer = nn.Linear(n_hidden, 1, F.relu)
    
    def forward(self, x: torch.Tensor):
        output = self.input_layer(x)
        output = self.hidden_layers(output)
        output = self.output_layer(output)
        return output
    
    def barf_parameters(self):
        return torch.cat([param.detach().flatten() for param in self.parameters()]).numpy()
```

## Convenience functions for training and testing

```python
def train_model(epochs, training_loader, validation_loader, LAMBD=0., nbatch=20, clip=False):
    LR = 1e-3
    # create our simple model
    model = SimpleModel(64, 3, 0.)
    params = model.parameters()
    # setup optimizer and loss functions
    optimizer = torch.optim.Adam(params, LR, weight_decay=LAMBD)
    loss_func = nn.MSELoss(reduction="mean")
    # track losses
    training_loss = torch.zeros(epochs)
    validation_loss = torch.zeros(epochs)
    # counters for training
    n_training = len(training_data)
    # tqdm gives a progress bar
    for epoch in tqdm(range(epochs)):
        epoch_loss = torch.zeros(len(training_loader))
        model.train()
        for index, (x, y) in enumerate(training_loader):
            optimizer.zero_grad()
            loss = loss_func(
                model(x.view(-1, 1)), y.view(-1, 1)
            )
            # perform backprop
            loss.backward()
            # update parameters
            optimizer.step()
            epoch_loss[index] = loss.detach()
        training_loss[epoch] = torch.mean(epoch_loss)
        # calculate testing loss
        epoch_validation_loss = torch.zeros(len(validation_loader))
        # set model to evaluation mode; this deactivates dropout
        model.eval()
        # run through the validation set
        with torch.no_grad():
            for index, (x, y) in enumerate(validation_loader):
                loss = loss_func(
                    model(x.view(-1, 1)), y.view(-1, 1)
                )
                # extract the loss
                epoch_validation_loss[index] = loss.item()
        # Calculate the loss epoch mean
        validation_loss[epoch] = torch.mean(epoch_validation_loss)
    return np.arange(epochs) + 1, training_loss, validation_loss, model
    
```

```python
def parameter_density_measurement(params, **kwargs):
    """
    This function is a convenience one, where we perform kernel density estimation on
    the flattened parameters. The `barf_parameters` class method of a model expels a
    flattened tensor of all the parameters of the model, and this in some sense is
    approximately the same as histogram counting all the different values.
    
    For now, the X range used to evaluate $\theta$ is preset: this is to ensure
    the entropy calculations are valid. That being said, I toyed with another idea
    of using quantile ranges to set the range to evaluate which I quite liked.
    """
    kde = KernelDensity(**kwargs)
    kde.fit(params.reshape(-1, 1))
#     first, third = np.quantile(params, [0.05, 0.995])
    X = np.linspace(-0.3, 0.3, 500)
    # KDE output of score_samples is log likelihood
    Y = np.exp(kde.score_samples(X.reshape(-1, 1)))
    # Make sure likelihood is normalized
    Y /= np.sum(Y)
    return X, Y
```

# Plotting relative entropy

This is an interesting way of looking at it, but I think I'll leave it for another article to elaborate upon. The idea is to compare the Shannon entropy of $p(\theta)$ with a uniform distribution; in the implementation below with `scipy`, this calculates the KL-divergence of our model from a uniform distribution (where $D_\mathrm{KL} = 0$ for the two to be equal).

```python
relative_H = entropy(np.tile(blank_y, (len(lambd_values), 1)),np.vstack(p_values), axis=1)
```

## Dropout regularization

This copies the earlier function, however replacing the hyperparameter $\lambda$ with the dropout probability $p$. The model is essentially the same as before, albeit with dropout layers now active rather than not.

```python
def train_dropout_model(epochs, training_loader, validation_loader, dropout_p, nbatch=20, clip=False):
    LR = 1e-3
    # create our simple model
    model = SimpleModel(64, 3, dropout_p)
    params = model.parameters()
    # setup optimizer and loss functions
    optimizer = torch.optim.Adam(params, LR)
    loss_func = nn.MSELoss(reduction="mean")
    # track losses
    training_loss = torch.zeros(epochs)
    validation_loss = torch.zeros(epochs)
    # counters for training
    n_training = len(training_data)
    # tqdm gives a progress bar
    for epoch in tqdm(range(epochs)):
        epoch_loss = torch.zeros(len(training_loader))
        model.train()
        for index, (x, y) in enumerate(training_loader):
            optimizer.zero_grad()
            loss = loss_func(
                model(x.view(-1, 1)), y.view(-1, 1)
            )
            loss.backward()
            # update parameters
            optimizer.step()
            epoch_loss[index] = loss.detach()
        training_loss[epoch] = torch.mean(epoch_loss)
        # calculate testing loss
        epoch_validation_loss = torch.zeros(len(validation_loader))
        model.eval()
        with torch.no_grad():
            for index, (x, y) in enumerate(validation_loader):
                loss = loss_func(
                    model(x.view(-1, 1)), y.view(-1, 1)
                )
                epoch_validation_loss[index] = loss.item()
        validation_loss[epoch] = torch.mean(epoch_validation_loss)
    model.eval()
    return np.arange(epochs) + 1, training_loss, validation_loss, model
```

## Run the dropout experiments

Run and retrain the model over a linearly spaced array of dropout $p$ values

```python
dropout_values = np.linspace(0., 1., 10)
results = list()
for p in dropout_values:
    results.append(
        train_dropout_model(50, train_loader, validation_loader, p)
    )
```

```python
def draw_neural_net(ax, left, right, bottom, top, layer_sizes):
    '''
    This code is taken from https://gist.github.com/craffel/2d727968c3aaebd10359
    
    I've made some small modifications to the code, namely the values of alpha
    used for drawing the edges.
    
    Draw a neural network cartoon using matplotilb.
    
    :usage:
        >>> fig = plt.figure(figsize=(12, 12))
        >>> draw_neural_net(fig.gca(), .1, .9, .1, .9, [4, 7, 2])
    
    :parameters:
        - ax : matplotlib.axes.AxesSubplot
            The axes on which to plot the cartoon (get e.g. by plt.gca())
        - left : float
            The center of the leftmost node(s) will be placed here
        - right : float
            The center of the rightmost node(s) will be placed here
        - bottom : float
            The center of the bottommost node(s) will be placed here
        - top : float
            The center of the topmost node(s) will be placed here
        - layer_sizes : list of int
            List of layer sizes, including input and output dimensionality
    '''
    n_layers = len(layer_sizes)
    v_spacing = (top - bottom)/float(max(layer_sizes))
    h_spacing = (right - left)/float(len(layer_sizes) - 1)
    # Nodes
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
        for m in range(layer_size):
            circle = plt.Circle((n*h_spacing + left, layer_top - m*v_spacing), v_spacing/4.,
                                color='w', ec='k', zorder=2,)
            ax.add_artist(circle)
    # Edges
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
        layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                line = plt.Line2D([n*h_spacing + left, (n + 1)*h_spacing + left],
                                  [layer_top_a - m*v_spacing, layer_top_b - o*v_spacing], c='k', alpha=0.1, zorder=1)
                ax.add_artist(line)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.axis('off')
draw_neural_net(ax, .1, .9, .1, .9, [1, 6, 6, 6, 1])
# Annotate the layers
ax.text(0.1, 0.4, "$X$", transform=ax.transAxes, zorder=5, horizontalalignment="center", verticalalignment="center")
ax.text(0.3, 0.1, "$h_1^{64}$", transform=ax.transAxes, zorder=5, horizontalalignment="center", verticalalignment="center")
ax.text(0.5, 0.1, "$h_2^{64}$", transform=ax.transAxes, zorder=5, horizontalalignment="center", verticalalignment="center")
ax.text(0.7, 0.1, "$h_3^{64}$", transform=ax.transAxes, zorder=5, horizontalalignment="center", verticalalignment="center")
ax.text(0.9, 0.4, "$f^*$", transform=ax.transAxes, zorder=5, horizontalalignment="center", verticalalignment="center")

fig.savefig('nn.png')
```
